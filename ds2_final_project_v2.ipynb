### Create Top 100 Book rating list

```python

# Load the CSV file
ratings_data = pd.read_csv('https://raw.githubusercontent.com/lilfakeS/schulich_data_science/main/Ratings.csv')

# Display the first few rows of the dataset to understand its structure
ratings_data.head()
```

```python
# Filtering out the books with non-zero ratings
non_zero_ratings_df = ratings_data[ratings_data['Book-Rating'] != 0]

# Grouping by ISBN and counting the number of non-zero ratings for each book
isbn_rating_counts = non_zero_ratings_df.groupby('ISBN')['Book-Rating'].count()

# Sorting the ISBNs by the number of non-zero ratings in descending order
sorted_isbn = isbn_rating_counts.sort_values(ascending=False)

# Selecting the top 100 unique ISBNs
top_100_isbn = sorted_isbn.head(100)
top_100_isbn.index.tolist()
```

```python
# Filtering the original dataset to keep only the rows with the top 100 ISBNs
top_100_isbn_list = top_100_isbn.index.tolist()
filtered_data = ratings_data[ratings_data['ISBN'].isin(top_100_isbn_list)]

# Display the first few rows of the filtered dataset
filtered_data.head()
```

### assign preference

```python
# Extracting unique genres from the top100_books_genres dataframe
unique_genres = set()
for genres in books_genres_df['Genre'].str.split(', '):
    unique_genres.update(genres)

# Adding a 'Genre Preferences' column to the selected_users dataframe
# Randomly assigning genres to users with the possibility of missing values
np.random.seed(0)  # For reproducibility
selected_users_df['Genre Preferences'] = selected_users_df.apply(
    lambda row: np.random.choice(
        list(unique_genres) + [np.nan], 
        size=np.random.randint(0, 3),  # Each user may have between 0 to 2 genre preferences
        replace=False
    ),
    axis=1
)

# Converting the array of genres to a comma-separated string
selected_users_df['Genre Preferences'] = selected_users_df['Genre Preferences'].apply(lambda x: ', '.join(x) if isinstance(x, np.ndarray) else x)
```

### Generate Browsing history

```python
# Setting a random seed for reproducibility
np.random.seed(0)

# Creating a browsing history dataframe
browsing_history = []

for _ in range(40000):
    user_id = np.random.choice(selected_users_df['User-ID'], replace=True)
    isbn = np.random.choice(filtered_books['ISBN'], replace=True)
    favorite = np.random.randint(0, 2)  # Randomly assign 0 or 1

    browsing_history.append([user_id, isbn, favorite])

browsing_history_df = pd.DataFrame(browsing_history, columns=['User-ID', 'ISBN', 'Favorite'])
```
