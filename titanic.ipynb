```python
titanic.drop('Name', axis=1, inplace=True)
titanic.drop('Cabin',axis= 1 , inplace = True)
```

```python
titanic.dropna(subset=['Embarked'], inplace= True)
```

```python
mean_age = titanic['Age'].mean()
```

```python
titanic['Age'].fillna(mean_age, inplace=True)
```

```python
titanic.describe(include='all')
```

```python
titanic.drop('PassengerId', axis=1, inplace=True)
```

```python
cat_columns = ['Pclass','Sex','Ticket', 'Embarked']
num_columns = ['Age','SibSp','Parch','Fare']
target = 'Survived'
```

```python
cat_transformer = OneHotEncoder(handle_unknown='ignore')
num_transformer = StandardScaler()
preprocessor = ColumnTransformer(transformers = [('cat',cat_transformer, cat_columns),
                                                 ('num', num_transformer, num_columns)])
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(titanic[cat_columns + num_columns], titanic['Survived'], test_size=0.3, random_state=111)
```

```python
X_transformed = preprocessor.fit_transform(titanic[cat_columns + num_columns])
X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.fit_transform(X_test)
```
