```python
# titles, authors, genres
titles = ["Shadows of Tomorrow", "Whispers of the Past", "Echoes of the Sea", "Dreams of Stardust",
          "Secrets of the Forest", "Legends of the Sky", "Mysteries of the Deep", "Glimpses of Dawn",
          "Fragments of the Night", "Journeys to the Unknown", "Tales of the Brave", "Chronicles of the Lost",
          "Sagas of the Wild", "Memories of the Ancient", "Visions of the Future", "Paths of Destiny",
          "Voices of the Wind", "Reflections of the Moon", "Guardians of the Realm", "Sorrows of the Heart",
          "Flames of Rebellion", "Songs of the Earth", "Dances with Shadows", "Traces of Light", "Echoes of Silence",
          "Whispers of Time", "Shadows of Fate", "Dreams Unseen", "Secrets Unspoken", "Legends Unwritten",
          "Mysteries Uncovered", "Glimpses Unveiled", "Fragments Unbound", "Journeys Uncharted", "Tales Untold",
          "Chronicles Unfolding", "Sagas Unraveling", "Memories Unearthed", "Visions Unleashed", "Paths Unveiled",
          "Voices Unheard", "Reflections Unseen", "Guardians Unleashed", "Sorrows Unshared", "Flames Unquenched",
          "Songs Unending", "Dances Unstoppable", "Traces Untraceable", "Echoes Unheard", "Whispers Unanswered"]
authors = ["Alexa Jordan", "Blake Morrison", "Charlie Phoenix", "Dakota Lee", "Evelyn Woods", "Finley Rivera",
           "Grayson Knight", "Harper Ellis", "Isaiah Clarke", "Jordan Hayes", "Kai Taylor", "Landon Carter",
           "Mason Brooks", "Nolan Foster", "Olivia Bennett", "Peyton Walsh", "Quinn Edwards", "Riley Adams",
           "Skyler Morgan", "Taylor Gibson", "Ulysses Grant", "Violet Shaw", "Weston Bailey", "Xander Reed",
           "Yasmin Ali", "Zachary Turner", "Avery Smith", "Bailey Johnson", "Carter Davis", "Danielle Brown"]
genres = ["Fantasy", "Science Fiction", "Mystery", "Thriller", "Romance", "Historical Fiction", "Young Adult"]

# Function to generate a random ISBN
def generate_isbn():
    return f'{random.randint(100, 999)}-{random.randint(100, 999)}-{random.randint(100, 999)}-{random.randint(10, 99)}'

# generating the dataset
data = []
for i in range(50):
    book = {
        "Title": random.choice(titles),
        "Author": random.choice(authors),
        "Genre": random.choice(genres),
        "Publication Year": random.randint(1980, 2023),
        "ISBN": generate_isbn(),
        "Fake Amazon Link": f"https://www.amazon.fake/book/{i+1}"
    }
    data.append(book)

df_books = pd.DataFrame(data)
df_books.head()

```

```python
df_books.to_csv('/Users/justinsheng/Downloads/books.csv', index=False)
```

```python
fake = faker.Faker()

# genres
genres = ["Fantasy", "Science Fiction", "Mystery", "Thriller", "Romance", "Historical Fiction", "Young Adult"]

# generating random genre preferences
def generate_genre_preferences():
    num_genres = random.randint(0, 3) 
    if num_genres == 0:
        return None  
    return ', '.join(random.sample(genres, num_genres))

# generating data for 100 users
data = []
for user_id in range(1, 101):
    user_data = {
        "User ID": user_id,
        "User Name": fake.user_name(),
        "Email": fake.email(),
        "Registration Date": fake.date_between(start_date="-5y", end_date="today"),
        "Genre Preferences": generate_genre_preferences()
    }
    data.append(user_data)

df_users = pd.DataFrame(data)
df_users.head()
```

```python
df_users.to_csv('/Users/justinsheng/Downloads/users.csv', index=False)
```

```python
# generating random ratings and reviews
def generate_rating_and_review():
    rating = random.randint(1, 5)

    if random.choice([True, False]):
        review = fake.sentence(nb_words=10)
    else:
        review = None  
    return rating, review

# generating 500 ratings and reviews again with the corrected function
data = []
for _ in range(500):

    user = df_users.sample().iloc[0]
    book = df_books.sample().iloc[0]


    rating, review = generate_rating_and_review()


    data.append({
        "User ID": user['User ID'],
        "User Name": user['User Name'],
        "Title": book['Title'],
        "Author": book['Author'],
        "Genre": book['Genre'],
        "Publication Year": book['Publication Year'],
        "ISBN": book['ISBN'],
        "Rating": rating,
        "Review": review
    })


df_ratings_reviews = pd.DataFrame(data)
df_ratings_reviews.head()
```

```python
df_ratings_reviews.to_csv('/Users/justinsheng/Downloads/ratings_reviews.csv', index=False)
```

```python
# generating random browsing time and favorite status
def generate_browsing_data():
    browsing_time = random.randint(1, 600)  # Random browsing time between 1 to 600
    favorite = random.choice([0, 1])  # Randomly choose 0 or 1 for favorite
    return browsing_time, favorite

# generating 500 browsing history entries
browsing_data = []
for _ in range(500):

    user = df_users.sample().iloc[0]
    book = df_books.sample().iloc[0]


    browsing_time, favorite = generate_browsing_data()


    browsing_data.append({
        "User ID": user['User ID'],
        "User Name": user['User Name'],
        "Title": book['Title'],
        "Author": book['Author'],
        "Genre": book['Genre'],
        "Publication Year": book['Publication Year'],
        "ISBN": book['ISBN'],
        "Favorite": favorite,
        "Browsing Time": browsing_time
    })


df_browsing_history = pd.DataFrame(browsing_data)
df_browsing_history.head() 
```

```python
# New set of numbers
new_numbers = [282, 4291, 263, 208, 74, 195, 68]

# New target number
new_target = 4568

# Using itertools.product to generate all combinations of +1 and -1
new_combinations = product([-1, 1], repeat=len(new_numbers))

# Attempting to find a combination that sums up to the new target
new_solution_found = False
for combo in new_combinations:
    result = sum(n * c for n, c in zip(new_numbers, combo))
    if abs(result - new_target) < 1e-6:  # Allowing a small error margin for floating point numbers
        new_solution = combo
        new_solution_found = True
        break

if new_solution_found:
    # Creating the equation from the found solution
    new_equation = " + ".join(f"{n * c}" for n, c in zip(new_numbers, new_solution) if c > 0)
    new_equation += " - " + " - ".join(f"{abs(n * c)}" for n, c in zip(new_numbers, new_solution) if c < 0)
    new_equation += f" = {new_target}"
else:
    new_equation = "No solution found with the given numbers."

new_equation

```
